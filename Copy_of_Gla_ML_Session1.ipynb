Calculate the avg marks and prin t class fialed or passed is the avg is 50 or above

```python
marks = [45, 60, 72, 38, 90]

total = 0

for i in range(len(marks) - 1):
    total += marks[i]

average = total / len(marks)

if average > 50:
    print("Class passed")
else:
    print("Class failed")

print("Average:", average)
```

```python
import pandas as pd
```

```python
data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "attendance_percent": [92, 67, 81, 45, 74, 88],
    "assignment_score": [18, 12, 15, 8, 14, 19],
    "quiz_score": [72, 48, 65, 30, 55, 80],
    "lab_completed": [True, False, True, False, True, True]
}

df = pd.DataFrame(data)

df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['total_score'] = df['assignment_score'] + df['quiz_score']
df
```

```python
df['eligible'] = (
    (df['attendance_percent'] >= 75) &
    (df['total_score'] >= 70) &
    (df['lab_completed'])
)
df
```

```python
eligible_students = df[df["eligible"] == True]
eligible_students
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['attendance_percent'], df['total_score'], color='blue', marker='o', s=100)

plt.title('Relationship Between Attendance and Total Score', fontsize=14)
plt.xlabel('Attendance Percent (%)', fontsize=12)
plt.ylabel('Total Score', fontsize=12)




plt.show()
```

```python



student_ids = df['student_id'].astype(str)


plt.bar(student_ids, df['total_score'], color='teal', edgecolor='black', width=0.6)

plt.title('Total Score achieved by each Student', fontsize=14)
plt.xlabel('Student ID')
plt.ylabel('Total Score')

plt.show()
```

```python



counts = df['eligible'].value_counts()


plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Eligible Students')


plt.show()
```
